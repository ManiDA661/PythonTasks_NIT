```python
import numpy as np
import matplotlib.pyplot as plt

list_PlName = ['Virat Kohli','Cheteswar Pujara','Michael Clarke','Virendra Sehwag','Mathew Hayden','Rahul Dravid','Ricky Ponting',
               'VVS Laxman','Sachin Tendulkar']
list_PlScore = [345,635,1606,1664,1888,2095,2141,2758,3158]

fig = plt.figure()
fig.set_size_inches(10,3)

plt.barh(list_PlName,list_PlScore,0.25,0.25,color='purple')

for i,v in enumerate(list_PlScore):
     plt.text(v,i,str(v),color='blue')
    
plt.xlabel(' No of Runs Scored in Border Gavaskar Series.',color='green',fontweight='bold')
plt.ylabel(' Players',color='green',fontweight='bold')
plt.title('Border Gavaskar Test Series',color='green',fontweight='bold')
plt.show()
```
